```python
import pandas as pd
import numpy as np
```

```python
countries = {"Poland":"PL",
            "Germany":"DE",
            "United Kingdom":"GB",
            "Austria":"AT",
            "Spain":"ES",
             "Portugal":"PT",
             "Netherlands":"NL",
            "Italy":"IT",
            "France":"FR",
            "Sweden":"SE",
            "Israel":"IL",
            "Belgium":"BG",
            "Portugal":"PT",
            "Netherlands":"NL",
            "Spain":"ES"}
cols = ['date','retail_and_recreation_percent_change_from_baseline',
       'grocery_and_pharmacy_percent_change_from_baseline',
       'parks_percent_change_from_baseline',
       'transit_stations_percent_change_from_baseline',
       'workplaces_percent_change_from_baseline',
       'residential_percent_change_from_baseline']
country_data = {}
for key,value in countries.items():
    data = pd.read_csv(f"2020_{value}_Region_Mobility_Report.csv")[cols]
    data = data.groupby("date").mean()
    data = data.rename({data.columns[0]:"mobility_recreation",data.columns[1]:"mobility_grocery",
                                          data.columns[2]:"mobility_parks",data.columns[3]:"mobility_transit",
                                          data.columns[4]:"mobility_work",data.columns[5]:"mobility_residential"},axis=1)
    data1 = pd.read_csv(f"2021_{value}_Region_Mobility_Report.csv")[cols]
    data1 = data1.groupby("date").mean()
    data1 = data1.rename({data1.columns[0]:"mobility_recreation",data1.columns[1]:"mobility_grocery",
                                          data1.columns[2]:"mobility_parks",data1.columns[3]:"mobility_transit",
                                          data1.columns[4]:"mobility_work",data1.columns[5]:"mobility_residential"},axis=1)
    full = data.append(data1)
    full["location"] = key
    country_data[key] = full
```

```python
data = country_data["Poland"].append(country_data["Germany"].append(country_data["Belgium"].append(country_data["France"].\
append(country_data["United Kingdom"].append(country_data["Sweden"].append(country_data["Israel"].append(country_data["Italy"].append(country_data["Spain"]\
        .append(country_data["Portugal"].append(country_data["Austria"].append(country_data["Netherlands"])))))))))))
```

```python
data.to_csv("mobility_report.csv")
```
